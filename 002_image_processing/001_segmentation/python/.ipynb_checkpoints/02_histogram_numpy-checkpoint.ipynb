```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

def main():
    img = cv.imread('images/T.jpg',0)

    plt.subplot(1,2,1)
    plt.imshow(img, cmap='gray')
    plt.title('Image')
    plt.xticks([])
    plt.yticks([])
    
    plt.subplot(1,2,2)
    hist,bins = np.histogram(img.ravel(),256,[0,255]) # ravel is flatten image
    plt.plot(hist)
    plt.title('histogram')
    plt.xlim(xmin=0,xmax=256)
    plt.ylim([0,3000])
    
    plt.show()
    
if __name__ == "__main__":
    main()
    
    
```
