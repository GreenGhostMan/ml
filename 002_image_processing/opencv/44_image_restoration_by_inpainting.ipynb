```python
import cv2 as cv
import matplotlib.pyplot as plt

def main():
    img = cv.imread('images/Damaged Image.tiff',1)
    mask= cv.imread('images/Mask.tiff',0)
    
    img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
    
    output1 = cv.inpaint(img, mask, 5, cv.INPAINT_TELEA)
    output2 = cv.inpaint(img, mask, 5, cv.INPAINT_NS)
    
    
    images = [img,mask,output1,output2]
    titles = ['Damaged', 'Mask', 'TELEA', 'NS']
    
    
    for i in range(4):
        plt.subplot(2,2,i+1)
        plt.imshow(images[i],cmap='gray')
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
        
    plt.show()
    
if __name__ == "__main__":
    main()
```
