```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('apple.jpg')
plt.imshow(img)
```

```python
gray = cv.imread('apple.jpg',0)
plt.imshow(gray)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(img)
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(gray)
```
