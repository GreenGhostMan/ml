```python
import cv2 as cv
import matplotlib.pyplot as plt


def main():
    img1 = cv.imread('lena_color_512.tif')
    img2 = cv.imread('house.tif')
    img1 = cv.cvtColor(img1,cv.COLOR_BGR2RGB)
    img2 = cv.cvtColor(img2,cv.COLOR_BGR2RGB)
    
    img3 = img1 * 1.5
    img4 = img2/1.5
    
    add = img1 + img2
    sub1= img1 - img2
    sub2= img2 - img1
    mul = img1 * img2
    lena50 = img1 + 50
    lena_50= img1 - 50
    fifty = 50 - img1
    division = img1/img2
    img3 = img1 * 1.5
    img4 = img2/1.5
    
    images = [img1, img2, add, sub1, sub2, mul, lena50, lena_50, fifty, division, img3, img4]
    titles = ['lena','house','lena+house', 'lena-house', 'house-lena', 'l*h','+50', '-50','50-lena','divide','*1.5','/1.5']
    
    for i in range(len(images)):
        plt.subplot(2,6,i+1)
        plt.imshow(images[i])
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
    
    plt.show()
    
    cv.waitKey(0)
    




if __name__ == "__main__":
    main()
```
