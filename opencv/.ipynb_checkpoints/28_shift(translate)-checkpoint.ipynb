```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def main():
    img = cv.imread('4.2.01.tiff')
    img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
    
    rows,clos,channels = img.shape
    
    T = np.float32([[1,0,50],[0,1,50]])
    print(T)
    
    output = cv.warpAffine(img,T,(clos,rows))
    
    plt.imshow(output)
    plt.title('Shifting')
    plt.show()


if __name__ == "__main__":
    main()
```
