```python
import cv2 as cv
import matplotlib.pyplot as plt

def main():
    img = cv.imread('lena_color_512.tif',0)

    plt.imshow(img)
    plt.show()

    plt.imshow(img, cmap="gray")
    plt.title("Gray")
    plt.show()

    plt.imshow(img, cmap='Spectral')
    plt.title("Spectral")
    plt.xticks([])
    plt.yticks([])
    plt.show()


if __name__ == "__main__":
    main()
```
