```python
import cv2 as cv
import matplotlib.pyplot as plt

def main():
    img = cv.imread('lena_color_512.tif')
    img2= cv.imread('lake.tif')
    
    img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
    img2= cv.cvtColor(img2,cv.COLOR_BGR2RGB)
    
    
    plt.subplot(1,2,1)
    plt.imshow(img)
    plt.title('lena')
    plt.xticks([])
    plt.yticks([])
    
    plt.subplot(1,2,2)
    plt.imshow(img2)
    plt.title('lake')
    plt.xticks([])
    plt.yticks([])
    
    plt.show()
    
     


if __name__ == "__main__":
    main()
```
