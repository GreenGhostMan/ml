```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def main():
    img = cv.imread('lake_color.tiff')
    
    img1 = cv.cvtColor(img,cv.COLOR_BGR2RGB)
    img2 = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
    
    img3 = abs(255-img1)
    img4 = abs(255-img2)
    
    images = [img1,img2,img3,img4]
    titles = ['Color','gray','Neg-Color','Neg-gray']
    
    for i in range(len(images)):
        plt.subplot(2,2,i+1)
        plt.imshow(images[i])
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
        
    plt.show()
    


if __name__ == "__main__":
    main()

```
