```python
import cv2 as cv
import matplotlib.pyplot as plt

def main():
    img = cv.imread('images/4.2.01.tiff')
    
    img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
    
    r,g,b = cv.split(img)
    
    images = [ cv.merge((r,g,b)),r,g,b]
    titles = ['img','r','g','b']
    cmap_values = ['plasma','Reds','Greens','Blues']
    
    for i in range(len(images)):
        plt.subplot(3,4,i+1)
        plt.imshow(images[i])
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
        
    for i in range(len(images)):
        plt.subplot(3,4,i+5)
        plt.imshow(images[i],cmap='gray')
        plt.title(titles[i]+"_cmap")
        plt.xticks([])
        plt.yticks([])
        
    for i in range(len(images)):
        plt.subplot(3,4,i+9)
        plt.imshow(images[i],cmap=cmap_values[i])
        plt.title(titles[i]+cmap_values[i])
        plt.xticks([])
        plt.yticks([])
        
    plt.show()



if __name__ == "__main__":
    main()
```
