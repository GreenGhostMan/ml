```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def main():
    img = cv.imread('images/4.2.01.tiff')
    img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
    
    row,col,channels = img.shape
    
    T = np.float32([[1,0,150],[0,1,20]])
    print(T)
    
    output = cv.warpAffine(img,T,(col,row))
    
    plt.imshow(output)
    plt.title('shifting')
    plt.show()


if __name__ == "__main__":
    main()
```
