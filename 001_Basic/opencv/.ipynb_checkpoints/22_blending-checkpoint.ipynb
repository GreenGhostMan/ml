```python
import cv2 as cv
import matplotlib.pyplot as plt


def main():
    img1 = cv.imread('images/lena_color_512.tif')
    img2 = cv.imread('images/house.tif')
    
    lena = cv.cvtColor(img1,cv.COLOR_BGR2RGB)
    house = cv.cvtColor(img2,cv.COLOR_BGR2RGB)
    
    alpha = 0.7
    beta = 0.3
    gamma = 0
    
    output1 = lena + house
    output2 = cv.addWeighted(lena,alpha,house,beta,gamma)
    
    images = [lena,house,output1,output2]
    titles = ['lena','house','output1','output2']
    
    for i in range(len(images)):
        plt.subplot(2,2,i+1)
        plt.imshow(images[i])
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])

    
    plt.show()
    
    cv.waitKey(0)
    




if __name__ == "__main__":
    main()
```

blending လုပ်ဖို့ alpha နဲ့ beta ပေါင်းခြင်းဟာ 1 ရရမှာဖြစ်ပြီး blendin လုပ်တဲ့သဘောဟာ image1 ကို alpha နဲ့ image2 ကို beta နဲ့ အသီးသီးမြှောက်ပီး အဲ့ဒီ ၂ ခုကိုပေါင်းပေးရမှာပါ။ ပြီးမှ gamma နဲ့ ထပ်ပေါင်းပေ့ါ။
