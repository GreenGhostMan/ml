```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

def main():
    img = cv.imread('images/4.2.01.tiff')
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    row,column,channel = img.shape
    
    point1 = np.float32([[0,0],[300,0],[0,300],[300,300]])
    point2 = np.float32([[0,0],[100,0],[0,100],[100,100]])
    #point2 = np.float32([[0,0],[500,0],[0,500],[500,500]])
    
    P = cv.getPerspectiveTransform(point1,point2)
    print(P)
    
    output = cv.warpPerspective(img, P, (500,500))
    
    plt.imshow(output)
    plt.title('Perspective tf')
    plt.show()

if __name__ == "__main__":
    main()
```
